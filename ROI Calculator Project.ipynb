```python
class ROI:

    def __init__(self, calculate_income):
            self.calculate_income = calculate_income


class Income(ROI):

    def __init__(self, calculate_income, income, laundry, storage, misc_income):
        self.income = income
        self.laundry = laundry
        self.storage = storage
        self.misc_income = misc_income
        ROI.__init__(self, calculate_income)

    def get_rent(self):
        while True:
            income = float(input("How much do you charge your tenants in rent per month? "))
            if income is not None:
                self.calculate_income += income
                laundry = float(input("How much do your tenants pay monthly for laundry? "))
            if laundry is not None:
                self.calculate_income += laundry
                storage = float(input("How much do you charge your tenants monthly for storage? "))
            if storage is not None:
                self.calculate_income += storage
                misc_income = float(input("How much do you charge your tenants for additional services? "))
            if misc_income is not None:
                self.calculate_income += misc_income
            print(f"Your income per month from the rental property is ${self.calculate_income:.2f}")
            return
            


# class ExpensesPerMonth(ROI):
#     def __init__(self, calculate_income, tax, insurance, utilities, other_expenses):
#         self.tax = tax
#         self.insurance = insurance
#         self.utilities = utilities
#         self.other_expenses = other_expenses
#         Income.__init__(self, calculate_income)

    def calculate_expenses(self, expenses):
        self.expenses = 0
        while True:
            tax = float(input("How much do you pay in taxes for this property? "))
            if tax is not None:
                self.expenses += tax
            insurance = float(input("How much do you pay insurance for this property? "))
            if insurance is not None:
                self.expenses += insurance
            utilities = float(input("How much do you pay for the utilities at this property? "))
            if utilities is not None:
                self.expenses += utilities
            other_expenses = float(input("If there are any other expenses that you cover for this property, how much do you pay? "))
            if other_expenses is not None:
                self.expenses += other_expenses
            print(f"Your monthly expenses come up to a total of ${self.expenses:.2f}. ")
            return


    def calculate_profit(self, monthly_profit):
        self.monthly_profit = self.calculate_income - self.expenses
        print(f"Your profit per month from your property comes up to ${self.monthly_profit:.2f}.")

    def total_investment(self, investment):
        self.investment = 0
        while True:
            down_payment = float(input("How much money did you put down when purchasing this property? "))
            if down_payment is not None:
                self.investment += down_payment
            closing_costs = float(input("How much did you spend on closing costs? "))
            if closing_costs is not None:
                self.investment += closing_costs
            rehab_budget = float(input("How much did you spend in repairs before putting your property up for rent?"))
            if rehab_budget is not None:
                self.investment += rehab_budget
            other_investments = float(input("If you've invested any additional money on this property, how much did you spend?"))
            if other_investments is not None:
                self.investment += other_investments
            print(f"You've invested a total of ${self.investment:.2f} on this rental property.")
            return

    def cash_on_cash(self, annual_cash):
        self.annual_cash = self.monthly_profit * 12
        print(f"Your annual cash flow is calculated to be {self.annual_cash:.2f}.")

    def total_roi(self, calc_investment, return_investment):
        self.calc_investment = self.annual_cash / self.investment
        self.return_investment = self.calc_investment * 10
        print("==================== RETURN ON INVESTMENT =========================")
        print(f"Your total return on investment on this property is {self.return_investment:.2f}%.")


i = Income(0,0,0,0,0)
i.get_rent()
i.calculate_expenses(0)
i.calculate_profit(0)
i.total_investment(0)
i.cash_on_cash(0)
i.total_roi(0,0)

# class ROI:
#     def calculate_return(self, down_payment, costs):
#         self.down_payment = down_payment
#         self.costs = costs
```
